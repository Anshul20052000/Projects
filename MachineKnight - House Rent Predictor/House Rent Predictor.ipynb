```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
columns = ['type','lease_type','furnishing','parking','facing','water_supply','building_type',
           'locality','LIFT','GYM','INTERNET','AC','CLUB','INTERCOM','POOL','CPA','FS','SERVANT',
           'SECURITY','SC','GP','PARK','RWH','STP','HK','PB','VP','Suburb','City_district','City',
           'County','State_district','State','Postcode']

encoded_values = {}

for col in columns:
    df[col]= encoder.fit_transform(df[col])
    encoded_values[col] = list(encoder.classes_)
```

```python
x = df.drop(["rent", "id"], axis=1)
x.head()
```

```python
y = df['rent']
```

```python
x = x.values
y = y.values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model=RandomForestRegressor(n_estimators=250, verbose=20)
```

```python
model.fit(x_train, y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
r2_score = model.score(x_test,y_test)
```

```python
print(r2_score*100,'%')
```

```python
import pickle

pickle.dump(model, open("Model_high_250_20.sav", "wb"))
```
