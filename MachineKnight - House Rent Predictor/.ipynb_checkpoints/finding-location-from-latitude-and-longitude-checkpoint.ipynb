```python
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent='geoapiExercises')
```

```python
geolocator.reverse("12.941603"+" , "+"77.568156").raw
```

```python
geolocator.reverse("12.941603"+" , "+"77.568156").raw['address']
```

```python
def location(cord):
    latitude = str(cord[0])
    longitude = str(cord[1])
    
    location = geolocator.reverse(latitude + " , " + longitude).raw['address']
    
    # If the values are missing then replace it with None
    
    if location.get('road') is None:
        location['road'] = None
        
    if location.get('suburb') is None:
        location['suburb'] = None
        
    if location.get('city_district') is None:
        location['city_district'] = None
        
    if location.get('city') is None:
        location['city'] = None
        
    if location.get('county') is None:
        location['county'] = None
        
    if location.get('state_district') is None:
        location['state_district'] = None
        
    if location.get('state') is None:
        location['state'] = None
        
    if location.get('postcode') is None:
        location['postcode'] = None
        
    if location.get('country') is None:
        location['country'] = None
        
    loc_update['Road'].append(location['road'])
    loc_update['Suburb'].append(location['suburb'])
    loc_update['City_district'].append(location['city_district'])
    loc_update['City'].append(location['city'])
    loc_update['County'].append(location['county'])
    loc_update['State_district'].append(location['state_district'])
    loc_update['State'].append(location['state'])
    loc_update['Postcode'].append(location['postcode'])
    loc_update['Country'].append(location['country'])
```

```python
loc_update = {"Road":[],
              "Suburb":[],
              "City_district":[],
              "City":[],
              "County":[],
              "State_district":[],
              "State":[],
              "Postcode":[],
              "Country":[]}
```

```python
import pickle

loc_update = {"Road":[],
              "Suburb":[],
              "City_district":[],
              "City":[],
              "County":[],
              "State_district":[],
              "State":[],
              "Postcode":[],
              "Country":[]}

for i, cord in enumerate(df.iloc[:,4:6].values):
    location(cord)
    #Continously Reading our Data and Saving it on the go !!
    pickle.dump(loc_update, open('train_loc_update.pickle', 'wb'))
    
    if i%100 == 0:
        print(i)

loc_update = {"Road":[],
              "Suburb":[],
              "City_district":[],
              "City":[],
              "County":[],
              "State_district":[],
              "State":[],
              "Postcode":[],
              "Country":[]}


# Testing Data
print("Testing Data")

for i, cord in enumerate(df2.iloc[:,4:6].values):
    location(cord)
    #Continously Reading our Data and Saving it on the go !!
    pickle.dump(loc_update, open('test_loc_update.pickle', 'wb'))
    
    if i%100 == 0:
        print(i)
```
